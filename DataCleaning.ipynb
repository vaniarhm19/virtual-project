#NEW CUSTOMER LIST

```python
import pandas as pd
import numpy as np
```

```python
nc_list = pd.read_excel('/content/KPMG_VI_New_raw_data_update_final.xlsx', sheet_name = 2)
```

```python
nc_list.head()
```

```python
nc_list.info()
```

```python
nc_list.duplicated()
```

```python
#Dropping the unnamed column
nc_list = nc_list.drop(['Unnamed: 16', 'Unnamed: 17', 'Unnamed: 18', 'Unnamed: 19', 'Unnamed: 20'], axis=1)
```

```python
#fill the null values in last_name
nc_list['last_name'] = nc_list['last_name'].fillna(value = nc_list['first_name'])
```

```python
#check the DOB
from pandas.api.types import is_datetime64_any_dtype
is_column_datetime_type = is_datetime64_any_dtype(nc_list['DOB'])
```

```python
#fill the null values in DOB
nc_list['DOB'] = nc_list['DOB'].fillna(np.random.choice(nc_list['DOB']))
```

```python
#fill the null values in job_title
nc_list['job_title'] = nc_list['job_title'].bfill(axis='rows')
```

```python
#fill the null values in job_industry_category
nc_list['job_industry_category'] = nc_list['job_industry_category'].bfill(axis='rows')
```

```python
#check the unique values of gender
np.unique(nc_list['gender'])
```

```python
#replace with new gender
nc_list['gender'] = nc_list['gender'].replace(['Male', 'Female'],['M', 'F'])
```

```python
#check the unique values of state
np.unique(nc_list['state'])
```

```python
#rounding the value in Value
nc_list['Value'] = nc_list['Value'].round(2)
```

```python
nc_list.tail()
```

```python
nc_list.info()
```

```python
from pandas.core.window.expanding import Axis
hasil = pd.DataFrame(data=nc_list)
file_name = "NewCustomer.csv"
hasil.to_csv(file_name)
print('DataFrame is written to CSV File successfully.')
```

#Customer Demographic

```python
import pandas as pd
import numpy as np
```

```python
CustDemo = pd.read_excel('/content/KPMG_VI_New_raw_data_update_final.xlsx', sheet_name = 3)
```

```python
CustDemo.head()
```

```python
CustDemo.info()
```

```python
#drop default column because it's unnecessary
CustDemo = CustDemo.drop('default', axis = 1)
```

```python
CustDemo.isnull().sum()
```

```python
#check duplicated id
CustDemo.duplicated('customer_id')
```

```python
#fill the null values in last_name
CustDemo['last_name'] = CustDemo['last_name'].fillna(value = CustDemo['first_name'])
```

```python
#fill the null values in job_title
CustDemo['job_title'] = CustDemo['job_title'].bfill(axis='rows')
```

```python
#fill the null values in job_industry_category
CustDemo['job_industry_category'] = CustDemo['job_industry_category'].bfill(axis='rows')
```

```python
#check the strange DOB
oldest = min(CustDemo['DOB'])
oldest
```

```python
#Drop the date
CustDemo = CustDemo.drop(CustDemo[CustDemo['DOB'] == '1843-12-21'].index)
```

```python
oldest1 = min(CustDemo['DOB'])
oldest1
```

```python
#fill the null values in DOB
CustDemo['DOB'] = CustDemo['DOB'].fillna(np.random.choice(CustDemo['DOB']))
```

```python
CustDemo.info()
```

```python
#fill the null values in tenure
mean_ten = CustDemo['tenure'].mean()
CustDemo['tenure'] = CustDemo['tenure'].fillna(value = mean_ten)
```

```python
#change the tenure type to int
CustDemo['tenure'] = CustDemo['tenure'].astype(int)
```

```python
CustDemo.head()
```

```python
CustDemo['gender'].unique()
```

```python
#replace with new gender
CustDemo['gender'] = CustDemo['gender'].replace(['Male', 'Female', 'Femal'],['M', 'F', 'F'])
```

```python
#replace with new value
CustDemo['owns_car'] = CustDemo['owns_car'].replace(['Yes', 'No'],['Y', 'N'])
```

```python
CustDemo.head()
```

```python
CustDemo.info()
```

```python
from pandas.core.window.expanding import Axis
hasil = pd.DataFrame(data=CustDemo)
file_name = "CustomerDemographic.csv"
hasil.to_csv(file_name)
print('DataFrame is written to CSV File successfully.')
```
